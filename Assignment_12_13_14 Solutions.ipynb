```python
while True :
    number = input("enter a positive integer number :")
    digits = len(number)
    sum_ = 0 
    
    if not number.isdigit() :
        print(number, "is invalid entry. Please enter valid input.")
        
    elif int(number) >= 0 :
        for i in range(len(number)) :
            sum_ = sum_ + int(number[i]) ** digits
            
        if sum_ == int(number) :
            print(number, " is an Amstrong Number")
            break
        else :
            print(number, " is not an Amstrong Number")
            break
    
```

<div class="alert alert-block alert-success">
<font color=red><b>Task :</b></font> 
<font color=purple><b>Write a program that takes a number from the user and prints the result to check if it is a</b></font> 
<font color=blue><b>prime number.</b></font>
<br><br>
<font color=MediumVioletRed>The examples of the desired output are as follows :<br></font>
<font color=magenta>input →  19 ; output : 19 is a prime number<br>
input →  10 ; output : 10 is not a prime number<br><br><font color=red>Note that; this question is famous on the web, so do it yourself to get more benefits from it. By the way, do not use try - except block.</font>

```python
n = int(input("Enter a positive integerr number to check if it is a Prime Number :"))

counter = 0

for i in range(1, n + 1):
    if n % i == 0 :
        counter += 1
        
if (n == 0) or (n == 1) or counter >= 3 :
    print(n, " is NOT a Prime Number")
    
else : 
    print(n, " is a Prime Number")
```

```python
n = int(input("Enter a positive integerr number to check if it is a Prime Number :"))
counter = 0

for i in range(2, n+1):
    if n % i == 0 :
        counter += 1
        
if counter == 1 :
    print("{} is a Prime Number".format(n))
    
else : 
    print("{} is NOT a Prime Number".format(n))
```

```python
def fibonacci(n):
    fibo = [1,1]
    
    for i in range(2, n):
        next_num = fibo[-1] + fibo[-2]
        fibo.append(next_num)
    return fibo

fibonacci(10)  

# fonksiyonum fibo listesini return etti (print etmedi!) 
# fakat jupyter, özelliği gereği, son return eden değeri ekranda gösterdi.
```
