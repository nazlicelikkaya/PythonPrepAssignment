## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Escape Sequences</p>

**Python official documentation : https://docs.python.org/3/reference/lexical_analysis.html**

**Escape sequence, özel bir işlevi olan karakterin önüne geldiğinde onun özelliğini ortadan kaldırır. onu sıradan bir karaketer haline dönüştürür**

```python
print('Ahmet Bey\'in ceketi')
```

```python
print('Ahmet Bey\in ceketi')
```

**Escape sequence, "n", "b" ve "t" karekterlerinin önüne geldiğinde onlara işlev ve özellik katar**

```python
print('Ayla\nın çantası')
```

```python
print('Ayla\\nın çantası')

# özel bir işlevi olan "\n" escape sequence'ının özelliğini 
# yine bir escape sequence ile kaldırdık
```

```python
print('C:\\north pole\noise_penguins.txt')
print('----------')
print('first', 'second', 'third', sep='\t')
```

```python
print('we are', '\boosting', 'our', '\brotherhood') 
print('it\'s essential to learn Python\'s libraries in IT World')
```
