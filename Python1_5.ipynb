## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Matter of Quotes</p>

**uzun bir text için 3 tırnak (3 çift veya 3 tek) kullanıyoruz**

```python
print("çift tırnakla başladım çift tırnakla bitirdim")
```

```python
print('tek tırnakla başladım tek tırnakla bitirdim')
```

```python
print("Hello World!')  

# çift tırnak açıldığı için kapatmak için yine çift tırnak bekliyor 
# ikinci çift tırnağı bulana kadar tüm karakterleri string olarak algılayacak!
```

```python
print("""benim adım Edwin, 
      ben sizi seviyorum
      sizde huzur buluyorum""")
```

```python
print("""hello world""")
```

```python
print('''hello world''')
```

```python
print("""123456789*-*/=-+.,^&/(+&/|~ef """)
```

```python
# birden fazla satırın veya uzun metinlerin içinde tek tırnak olabilir.
# kesme "/" olabilir.
# I am derken I'dan sonra metnin içinde üstten virgül gelebilir.
# metnin içinde çift tırnak " kullanmak isteyebilirsin.

# bütün bu durumlarda hata almamak için uzun metinlerde 3'lü tırnak kullanırız.
```

```python
print("Ahmet Bey'in kedisi")
```

**Cümle içinde tek tırnak hatalı kullanıma örnek :**

```python
print('Sevim'in kedisi')
```

```python
print("Sevim'in kedisi")
```

```python
print('Sevim"in kedisi')
```

**Hatalı bir tırnak açma kapamaya örnek :**

```python
print()
print('''')
```

```python
# boş string (aç-kapa tırnak)

print('')
```
