## Virtual environment'in oluştuğunu nasıl anlarız?

**Sanal ortamı etkinleştirdikten sonra environment ismi, terminal prompt'un  başlangıcında parantez içinde görüntüleniyor. Böylece virtual environmentimizin aktif olduğunu anlayabiliyoruz.**

**Bunu anlamanı bir başka yöntemi: windows sistemi için "where python", Linux sistemleri için "which python" komutunu çalıştırmaktır.** 

**Bu komutu çalıştırırsak virtual environment içindeki Python yorumlayıcısının konumunu gösteriyor.**

```python
where python   # {windows için}

# which python  # {Linux sistemleri için}
```

**Başta az önce oluşturduğumuz environment klasörü içinde olmak üzere localdeki diğer tüm python.exe dosyalarının yerini görebiliyoruz.**

**NOT: Conda environment'i tespit edemediğinden, conda kurulu olsa bile conda environment'ının python.exe dosyasını gösteremez. Bununla birlikte Anaconda, diğer tüm environmentlerı da tespit edebilir.**

**Ayrıca sanal ortamın Python sürümünü kontrol ederek de sanal ortamın oluştuğunu anlayabiliriz.**

**Çünkü oluşturduğumuz sanal ortamdaki python sürümü, bu sanal ortamı kurmak için kullandığımız Python sürümü ile aynıdır.**

```python
python —V

# veya: 
python --version
```

### <p style="background-color:#117A65; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Environment'a package yükleme & "pip" komutları</p>

```python
python -m pip install --upgrade pip  # {windows için}

# python3 -m pip install --upgrade pip # {linux sistemleri için}
```

**Dikkat! Sanal ortamda çalıştığımız için bu komut yalnızca aktif olan sanal ortamdaki pip tool'unu upgrade edecektir.**

**Bir paketi kaldırmak için:**

### <p style="background-color:#117A65; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Reproducing işlemi</p>

**Bir iş arkadaşınızın sizdeki aynı environment'i oluşturabilmesi için; projenin sanal ortamınızda yüklü olan tüm "dependency"leri listelemek ve bu listeyi (aynı içerikleri) arkadaşınıza göndermeniz gerekiyor.** 

**Yüklü paketleri gereksinimler biçiminde output almak için:**

```python
pip freeze

# {çalıştır}
```

**Sonraki adımda bu paket listesini "requirements.txt" dosyasına aktarıp txt dosyası olarak iş arkadaşınıza gönderin:**

```python
# listeyi current folder'da "requirements.txt" dosyasına kaydetmek için.

pip freeze > requirements.txt  
```

```python
notepad requirements.txt  # {windows için.. notepad'i açarak dosya içeriğini gösterir.}

cat requirements.txt  # {linux sistemleri için dosya içeriğini gösterir} 
```

**İş arkadaşınızın yapacağı işlemler:**

```python
python -m venv myfriend_env

.\myfriend\Scripts\activate

# gerekli tüm paketleri yüklemek için: 
pip install -r requirements.txt
```

**Projeyi bir Git repository'ye ekleyecekseniz, virtual environment klasörünü repository'e eklemeyin. Sadece requirements.txt dosyasını ekleyin.**

### <p style="background-color:#117A65; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Deaktivasyon ve silme işlemleri</p>

**Virtual environment üzerinde çalışmayı bitirdikten sonra veya başka bir virtual environmenta geçmek istediğinizde mevcut virtual environment'imizi devre dışı bırakmak için sanal ortamı deaktif hale getiriyoruz:**

```python
deactivate
```

**Bir sanal ortamı silmek için; basitçe onun klasörünü siliyoruz:**

```python
rmdir /s projem  # {windows için}

rm -rf projem  # {linux için}

# rmdir /s projem   SADECE ENV KLASÖRÜNÜ SİLMEK İÇİN
# rm -rf projem/projem_env SADECE ENV KLASÖRÜNÜ SİLMEK İÇİN
```

**rmdir: Deletes a directory.** 

**/s	Belirtilen dizin ve tüm dosyalar dahil tüm alt dizinleri siler.**

**/q	Sessiz modu belirtir. Bir klasörü silerken onay istemez.**

**(/q parametresi sadece /s parametresi kullanıldığında iş görür)**

**DİKKAT: /q modu çalıştırdığınızda tüm klasör onaylanmadan silineceği için bu seçeneği kullanmadan önce önemli dosyaların taşındığından veya yedeklendiğinden emin olmak gerekir.**

## Alternatif environment oluşturma komutları:

#### Sadece bir environment oluşturmak için:

```python
conda create --name ENV_NAME
```

#### Belirli bir Python sürümüyle bir ortam oluşturmak için:

```python
conda create -n ENV_NAME python=3.9
```

#### Belirli bir pakete sahip bir ortam oluşturmak için:

```python
conda create -n ENV_NAME scipy

veya:
    
conda create -n ENV_NAME python
conda install -n ENV_NAME scipy
```

#### Bir paketin belirli bir sürümüne sahip bir ortam oluşturmak için:

```python
conda create -n ENV_NAME scipy=0.17.3
```

#### Python'un belirli bir sürümü ile birlikte birden çok paketi olan bir ortam oluşturmak için:

```python
conda create -n ENV_NAME python=3.9 scipy=0.17.3 seaborn pandas
```
