# seek() method 

**Change stream position.**

**Change the stream position to the given byte offset. The offset is
interpreted relative to the position indicated by whence.**

```python
with open("fishes.txt", "r", encoding = "utf-8") as file :
    print(file.read(33))
    print(file.read(33))
    
    file.seek(0)
    print(file.read(33))
```

# tell() method

**Return current stream position**

```python
with open("fishes.txt", "r", encoding = "utf-8") as file :
    print(file.read(33))
    print(file.read(33))
  
    file.seek(0)
    
    print(file.read(33))
    print(file.tell())
```

```python
with open("fishes.txt", "r", encoding = "utf-8") as file :
    print(file.read(33))
    print(file.read(15))
    print(file.tell())
    
    file.seek(0)
    print(file.read(33))
    print(file.tell())
```

**Task: Put a space between the stanzas of the poem**

```python
counter = 0

with open("istiklal.txt", "w", encoding="utf-8") as file:
    
    for i in lines:
        counter +=1
        if counter % 4 == 0 :
            file.write(i + "\n")
            
        else:
            file.write(i)
```
